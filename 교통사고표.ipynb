```python
import matplotlib.pyplot as plt
import seaborn as sb

plt.figure(figsize=(15, 5))
plt.xticks(rotation=60)

df_sorted['부상자수'] = df_sorted['부상자수'].fillna('0').str.replace(',', '').str.extract('(\d+)').astype(int)  # NaN 값을 0으로 대체하고 숫자로 변환합니다.
df_sorted = df_sorted.sort_values('부상자수', ascending=False)  # '부상자수'를 내림차순으로 정렬합니다.

sb.pointplot(data=df_sorted, x='년도', y='부상자수')

plt.show()

```

```python
plt.figure(figsize=(15,5))
plt.xticks(rotation=60)


sb.pointplot(data=df_sorted, x='년도', y='사망자수')
sb.pointplot(data=df_sorted, x='년도', y='부상자수',color ='red')
sb.pointplot(data=df_sorted, x='년도', y='사고건수',color ='green')
```

```python
import matplotlib.pyplot as plt
import seaborn as sb

fig, ax1 = plt.subplots(figsize=(15, 5))
plt.xticks(rotation=60)

# 사망자수에 대한 막대 그래프 그리기
sb.barplot(data=df_sorted, x='년도', y='사망자수', ax=ax1, color='red')

# 사망자수의 단위를 오른쪽에 표시하기
ax2 = ax1.twinx()
ax2.set_ylabel('사망자수', color='red')
ax2.tick_params(axis='y', labelcolor='red')

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sb

plt.figure(figsize=(15, 5))
plt.xticks(rotation=60)

# 사망자수에 대한 막대 그래프 그리기
ax1 = sb.barplot(data=df_sorted, x='년도', y='사망자수', color='red')

# 사망자수의 단위를 오른쪽에 표시하기
ax2 = ax1.twinx()
ax2.set_ylabel('사망자수', color='black')
ax2.tick_params(axis='y', labelcolor='black')

# 부상자수에 대한 점 그래프 그리기
sb.pointplot(data=df_sorted, x='년도', y='부상자수', color='black')

# 사고건수에 대한 점 그래프 그리기
sb.pointplot(data=df_sorted, x='년도', y='사고건수', color='green')

plt.show()
```

```python
plt.figure(figsize=(15, 5))
plt.xticks(rotation=60)

sb.pointplot(data=df_sorted, x='년도', y='사망자수')
sb.pointplot(data=df_sorted, x='년도', y='부상자수', color='red')
sb.pointplot(data=df_sorted, x='년도', y='사고건수', color='green')

plt.ylabel('사망자수 / 부상자수 / 사고건수')

plt.show()
```
